```python
# define required input files
# import dataFrame with calculated results from Tunnel GIS
solar_data_df = pd.read_csv('SolarIrradiation.NT1.2004.nasa.csv',index_col=0)
```

```python
irr = solar_data_df['swv_dwn'].tolist()

trace = go.Scatter(
        y=signal.savgol_filter(irr, 53, 3),  #https://plot.ly/python/smoothing/
        x=arange(1, 365, 1),
        name= 'kWh/m2',
        )

#print(trace)

layout=go.Layout(title="Horizontal Solar Irradiation - NT1 area", 
                 xaxis={'title':'day'},
                 yaxis={'title':'kWh/m2/day'},
                 width=700,
                 height=400,
                )

figure=go.Figure(data=[trace],layout=layout)
plotly.iplot(figure)
```
